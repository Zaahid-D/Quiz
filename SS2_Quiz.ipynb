```python
import random

#Quiz Questions Data Structure

quiz_questions = [
    {
        'category': 'History',
        'difficulty_level' : 'Beginner',
        'question' : 'What year did South Africa gain independence from Britain?',
        'options' : ['A. 1910', 'B. 1945', 'C. 1961', 'D. 1994'],
        'correct_answer' : 'C'
    },
    {
        'category': 'History',
        'difficulty_level': 'Beginner',
        'question': 'What was the main export of the Mali Empire?',
        'options': ['A. Gold', 'B. Salt', 'C. Ivory', 'D. Diamonds'],
        'correct_answer': 'A'
    },
    {
        'category': 'History',
        'difficulty_level': 'Beginner',
        'question': 'The Great Zimbabwe was a city in the historical African kingdom of __________.',
        'options': ['A. The Kingdom of Mapungubwe', 'B. The Kingdom of Zimbabwe', 'C. The Kingdom of Mutapa', 'D. The Kingdom of Butua'],
        'correct_answer': 'B'
    },
    {
        'category': 'History',
        'difficulty_level': 'Moderate',
        'question': 'Kwame Nkrumah was the first president of Ghana.',
        'options': ['A. True', 'B. False'],
        'correct_answer': 'A'
    },
    {
        'category': 'History',
        'difficulty_level': 'Moderate',
        'question': 'The Battle of Adwa was a victory for which African country against Italy in 1896?',
        'options': ['A. Ethiopia', 'B. Eritrea', 'C. Somalia', 'D. Sudan'],
        'correct_answer': 'A'
    },
    {
        'category': 'History',
        'difficulty_level': 'Moderate',
        'question': 'The Ashanti Empire was located in which modern-day country?',
        'options': ['A. Nigeria', 'B. Ghana', 'C. Ivory Coast', 'D. Burkina Faso'],
        'correct_answer': 'B'
    },
    {
        'category': 'History',
        'difficulty_level': 'Advanced',
        'question': 'The Kingdom of Kush was located in modern-day __________.',
        'options': ['A. Egypt', 'B. Sudan', 'C. Ethiopia', 'D. Libya'],
        'correct_answer': 'B'
    },
    {
        'category': 'History',
        'difficulty_level': 'Advanced',
        'question': 'Cetshwayo kaMpande led the Zulu Kingdom during the Anglo-Zulu War.',
        'options': ['A. True', 'B. False'],
        'correct_answer': 'A'
    },
    {
        'category': 'History',
        'difficulty_level': 'Advanced',
        'question': 'The Axum Empire, one of the four greatest civilizations in the world, was located in which modern-day country?',
        'options': ['A. Egypt', 'B. Ethiopia', 'C. Sudan', 'D. South Africa'],
        'correct_answer': 'B'
    },
    {
        'category': 'History',
        'difficulty_level': 'Challenging',
        'question': 'Arrange the following empires in order of their historical emergence, from earliest to latest: Mali Empire, Kingdom of Kush, Songhai Empire, and Kingdom of Benin',
        'options': [
            'A. Mali Empire, Kingdom of Kush, Kingdom of Benin, Songhai Empire',
            'B. Kingdom of Kush, Kingdom of Benin, Mali Empire, Songhai Empire',
            'C Kingdom of Benin, Kingdom of Kush, Songhai Empire, Mali Empire',
            'D Songhai Empire, Kingdom of Benin, Mali Empire, Kingdom of Kush'
        ],
        'correct_answer': 'B'
    },
    {
        'category': 'History',
        'difficulty_level': 'Challenging',
        'question': 'Identify the correct statement regarding the impact of colonialism on African societies:',
        'options': [
            'A. Colonialism led to the preservation of indigenous cultures and traditions.',
            'B. African societies experienced rapid industrialization under colonial rule.',
            'C. Colonial borders often divided ethnic groups, leading to conflicts that persist to this day.',
            'D. Colonial powers provided equal rights and opportunities to all Africans.'
        ],
        'correct_answer': 'C'
    },
    {
        'category': 'History',
        'difficulty_level': 'Challenging',
        'question': 'Fill in the blank with the correct term: During the ________________, European powers met to carve up Africa, leading to the establishment of colonial borders that often disregarded indigenous territories and ethnic groups.',
        'options': ['A. Slave Trade', 'B. Scramble for Africa', 'C. Great Migration', 'D. Pan-African Congress'],
        'correct_answer': 'B'
    },

    {
        'category': 'Culture',
        'difficulty_level': 'Beginner',
        'question': 'What is the traditional religion of the Yoruba people?',
        'options': ['A. Christianity', 'B. Islam', 'C. Yoruba religion', 'D. Hinduism'],
        'correct_answer': 'C'
    },
    {
        'category': 'Culture',
        'difficulty_level': 'Beginner',
        'question': 'The Maasai people are native to Kenya and Uganda.',
        'options': ['A. True', 'B. False'],
        'correct_answer': 'B'
    },
    {
        'category': 'Culture',
        'difficulty_level': 'Beginner',
        'question': 'What is the traditional method of communication among the San people of Southern Africa?',
        'options': ['A. Drumming', 'B. Smoke signals', 'C. Petroglyphs', 'D. Oral storytelling'],
        'correct_answer': 'D'
    },
    {
        'category': 'Culture',
        'difficulty_level': 'Moderate',
        'question': 'The Dogon people, known for their astronomical knowledge, are from which country?',
        'options': ['A. Mali', 'B. Burkina Faso', 'C. Niger', 'D. Chad'],
        'correct_answer': 'A'
    },
    {
        'category': 'Culture',
        'difficulty_level': 'Moderate',
        'question': 'The Adinkra symbols originate from which African culture?',
        'options': ['A. Ashanti', 'B. Zulu', 'C. Maasai', 'D. Yoruba'],
        'correct_answer': 'A'
    },
    {
        'category': 'Culture',
        'difficulty_level': 'Moderate',
        'question': 'The Kente cloth is native to which African country?',
        'options': ['A. Nigeria', 'B. Ghana', 'C. Senegal', 'D. Ivory Coast'],
        'correct_answer': 'B'
    },
    {
        'category': 'Culture',
        'difficulty_level': 'Advanced',
        'question': 'The Griot tradition of storytelling is most associated with East Africa.',
        'options': ['A. True', 'B. False'],
        'correct_answer': 'B'
    },
    {
        'category': 'Culture',
        'difficulty_level': 'Advanced',
        'question': 'What is the name of the traditional Ethiopian communal serving dish often used in traditional meals?',
        'options': ['A. Tagine', 'B. Injera', 'C. Jollof', 'D. Ugali'],
        'correct_answer': 'B'
    },
    {
        'category': 'Culture',
        'difficulty_level': 'Advanced',
        'question': 'The Sankofa symbol, which represents the importance of learning from the past, originates from which African culture?',
        'options': ['A. Zulu', 'B. Maasai', 'C. Ashanti', 'D. Yoruba'],
        'correct_answer': 'C'
    },
    {
        'category': 'Notable Figures',
        'difficulty_level': 'Beginner',
        'question': 'Who is the South African anti-apartheid revolutionary who served as President of South Africa from 1994 to 1999?',
        'options': ['A. Desmond Tutu', 'B. F. W. de Klerk', 'C. Nelson Mandela', 'D. Thabo Mbeki'],
        'correct_answer': 'C'
    },
    {
        'category': 'Notable Figures',
        'difficulty_level': 'Beginner',
        'question': 'Ellen Johnson Sirleaf was the first female president in Africa.',
        'options': ['A. True', 'B. False'],
        'correct_answer': 'A'
    },
    {
        'category': 'Notable Figures',
        'difficulty_level': 'Beginner',
        'question': 'Christiaan Barnard performed the world’s first successful heart transplant.',
        'options': ['A. True', 'B. False'],
        'correct_answer': 'A'
    },
    {
        'category': 'Notable Figures',
        'difficulty_level': 'Moderate',
        'question': 'Who was the first black Nobel laureate?',
        'options': ['A. Nelson Mandela', 'B. Martin Luther King Jr.', 'C. Albert Luthuli', 'D. Desmond Tutu'],
        'correct_answer': 'C'
    },
    {
        'category': 'Notable Figures',
        'difficulty_level': 'Moderate',
        'question': 'Chinua Achebe is the author of “__________”.',
        'options': ['A. Things Fall Apart', 'B. The Lion and the Jewel', 'C. Anthills of the Savannah', 'D. Arrow of God'],
        'correct_answer': 'A'
    },
    {
        'category': 'Notable Figures',
        'difficulty_level': 'Moderate',
        'question': 'Who is the Kenyan woman who won the Nobel Peace Prize in 2004?',
        'options': ['A. Wangari Maathai', 'B. Ellen Johnson Sirleaf', 'C. Joyce Banda', 'D. Ameenah Gurib'],
        'correct_answer': 'A'
    },
    {
        'category': 'Notable Figures',
        'difficulty_level': 'Advanced',
        'question': 'Who was the Ghanaian independence leader and the first Prime Minister and President of Ghana?',
        'options': ['A. Patrice Lumumba', 'B. Kwame Nkrumah', 'C. Jomo Kenyatta', 'D. Julius Nyerere'],
        'correct_answer': 'B'
    },
    {
        'category': 'Notable Figures',
        'difficulty_level': 'Advanced',
        'question': 'Desmond Tutu is the South African social rights activist and retired Anglican bishop who won the Nobel Peace Prize in __________.',
        'options': ['A. 1980', 'B. 1984', 'C. 1990', 'D. 1994'],
        'correct_answer': 'B'
    },
    {
        'category': 'Notable Figures',
        'difficulty_level': 'Advanced',
        'question': 'Ken Saro-Wiwa was the Nigerian environmental activist who led the Movement for the Survival of the Ogoni People (MOSOP).',
        'options': ['A. True', 'B. False'],
        'correct_answer': 'A'
    },
]

#Implement Functions
def generate_question(category, difficulty):
    """
    Generate a random question based on the specified category and difficulty level.
    """
    filtered_questions = []
    for q in quiz_questions:
        if q['category'] == category and q['difficulty_level'] == difficulty:
            filtered_questions.append(q)
    return random.choice(filtered_questions)

def validate_answer(question, player_answer):
    """
    Check if the player's answer is correct by comparing it with the 'correct_answer' key in the question dictionary.
    """
    return player_answer == question['correct_answer']

def update_score(score, is_correct):
    """
    Update the player's score based on the correctness of the answer.
    """
    if is_correct:
        return score + 1
    else:
        return score

def get_player_answer(question):
    """
    Display the question and options to the player as well as get the player's answer input.
    """
    print(f"Category: {question['category']} | Difficulty Level: {question['difficulty_level']}")
    print(question['question'])
    for option in question['options']:
        print(option)
    
    # Loop until the player provides a valid single-letter answer
    while True:
        player_answer = input("Please enter the letter corresponding to your choice: ").strip().upper()
        if len(player_answer) == 1 and player_answer.isalpha():
            return player_answer
        else:
            print("The input you've provided is in the wrong format. Please enter only the letter corresponding to the option.")

#Implement Quiz Loop
def play_quiz(total_levels):
    """
    Play the quiz game with specified levels and categories.
    """
    # Welcome message and unique username
    names = ['HeritageHunter', 'MandelaMind', 'NileNavigator', 'BaobabBrain', 'MaliMystic']
    username = random.choice(names) + str(random.randint(100, 999))
    print("Welcome to the African Trivia Quiz!")
    print(f"Hello, {username}! Let's test your knowledge on Africa. Are you ready to play?")
    print("You will go through 10 levels of questions, each increasing in difficulty. The questions will cover three categories: History, Culture, and Notable Figures. Let's begin!\n")

    level = 1
    score = 0
    while level <= total_levels:
        print(f"Level {level}")
        if level == 1:
            question = generate_question('History', 'Beginner')
        elif level == 2:
            question = generate_question('Culture', 'Beginner')
        elif level == 3:
            question = generate_question('Notable Figures', 'Beginner')
        elif level == 4:
            question = generate_question('History', 'Moderate')
        elif level == 5:
            question = generate_question('Culture', 'Moderate')
        elif level == 6:
            question = generate_question('Notable Figures', 'Moderate')
        elif level == 7:
            question = generate_question('History', 'Advanced')
        elif level == 8:
            question = generate_question('Culture', 'Advanced')
        elif level == 9:
            question = generate_question('Notable Figures', 'Advanced')
        elif level == 10:
            question = generate_question('History','Challenging')
            
        player_answer = get_player_answer(question)
        is_correct = validate_answer(question, player_answer)
        score = update_score(score, is_correct)
        if is_correct:
            congratulatory_messages = [
                f"Bravo, {username}! You've nailed it!",
                f"Outstanding, {username}! Your answer is spot on!",
                f"You're on fire, {username}! That's the right answer!",
                f"Fantastic job, {username}! You've chosen wisely!",
                f"Incredible, {username}! You've cracked it!"
            ]
            print(random.choice(congratulatory_messages))
        else:
            incorrect_messages = [
                f"Oops, {username}! That's not quite right. The correct answer is:",
                f"Better luck next time, {username}! The correct answer is:",
                f"Incorrect, {username}. The correct answer is:",
                f"Not quite there, {username}. The correct answer is:",
                f"Almost, {username}! The correct answer is:"
            ]
            incorrect_message = random.choice(incorrect_messages)
            correct_option = question['correct_answer']
            print(f"{incorrect_message} {correct_option}")
        
        print(f"Your current score is {score}/10")
        print()
        level += 1

    print("Quiz completed!")
    print(f"Final Score: {score}/{total_levels}")
    # Evaluate final score and provide feedback
    if score <= 3:
        print(f"{username}, you have very basic knowledge. Keep learning and improving!")
    elif score <= 6:
        print(f"{username}, you have a moderate level of knowledge. Good job!")
    elif score <= 8:
        print(f"{username}, you have a good understanding of the material. Well done!")
    elif score <= 9:
        print(f"{username}, you have an excellent understanding of the material. Great work!")
    else:
        print(f"Perfect score, {username}! You have outstanding knowledge!")


# Set total_levels and call play_quiz to run the quiz game
total_levels = 10
play_quiz(total_levels)
```
